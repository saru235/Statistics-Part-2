 **1.What is hypothesis testing in statistics?**

Hypothesis testing is a fundamental statistical method used to make inferences or decisions about a population based on sample data by evaluating if there is enough evidence to support a specific claim (hypothesis).

**- Overview of Hypothesis Testing**

Hypothesis testing involves formulating two competing claims about a population parameter:

** - Null Hypothesis (H₀):** The default or no-effect assumption, e.g., there is no difference or no effect.

**- Alternative Hypothesis (H₁ or Hₐ):** The claim you want to test or prove, representing an effect or difference.
The process uses sample data to assess whether to reject the null hypothesis or fail to reject it based on the strength of the evidence provided by the data.

**2.What is the null hypothesis, and how does it differ from the alternative hypothesis?**

- The **null hypothesis** (H₀) is a statement that there is no effect, relationship, or difference between variables or groups. It typically reflects the status quo or a position of no influence.

- For example, if evaluating a new drug, the null hypothesis might state, “the drug has no effect on patient recovery rates.”

- In mathematical terms, it often involves an equality (e.g., H₀: µ = 50, meaning the population mean is equal to 50)

**3.What is the significance level in hypothesis testing, and why is it important?**

The significance level is the threshold probability set by a researcher to decide when to reject the null hypothesis; it represents the risk of falsely rejecting a true null hypothesis, making it a fundamental component in statistical decision making.

**- What is the Significance Level?**

The significance level, often denoted by α, is a predefined cutoff that sets the maximum acceptable probability of making a Type I error in hypothesis testing. A Type I error occurs when the null hypothesis (typically representing no effect or no difference) is wrongly rejected despite being true. Common choices for α are 0.05, 0.01, or 0.10, meaning there is a 5%, 1%, or 10% risk respectively of rejecting the null hypothesis incorrectly.

**- Why is the Significance Level Important?**

**- Controls Error Rate:** The significance level directly controls the chance of making a false positive conclusion, which helps maintain scientific rigor.

**- Decision Framework:** It provides a clear rule for deciding whether observed data is sufficiently unlikely under the null hypothesis to warrant rejection.

**- Standardization:** Using conventional α values (like 0.05) allows for consistent interpretation and comparison of results across studies.

**- Balance of Risks: **Choosing α balances the risk of Type I errors against Type II errors (failing to detect a true effect), helping tailor tests to the needs of the context.

**4.What does a P-value represent in hypothesis testing?**

A P-value measures the probability of obtaining test results at least as extreme as the observed data, assuming the null hypothesis is true, helping decide whether to reject the null hypothesis.

**- Definition and Role**

In hypothesis testing, the P-value quantifies how compatible your observed data is with a stated null hypothesis. Specifically, it represents the probability of observing a result as extreme or more extreme than what was actually observed, assuming the null hypothesis is true. A smaller P-value indicates stronger evidence against the null hypothesis.

**- Interpretation in Decision-Making**

If the **P-value is less than or equal to a pre-determined significance level **(commonly 0.05), it suggests that such extreme observed data would be unlikely under the null hypothesis, leading to rejection of the null hypothesis.

If the** P-value is greater than the significance level**, there is insufficient evidence to reject the null hypothesis, meaning the observed data is consistent with it.

**5.How do you interpret the P-value in hypothesis testing?**

**- How to Interpret the P-value**

   - A **small P-value (typically ≤ 0.05)** indicates that the observed data is unlikely under the null hypothesis. This level of rarity suggests that the null hypothesis may not hold, and **you reject the null hypothesis** in favor of the alternative hypothesis.
     
  - A **large P-value (typically > 0.05)** suggests the observed data is consistent with the null hypothesis. There is insufficient evidence to reject H0, so you **fail to reject the null hypothesis**, but it does not prove the null is true.

**6.What are Type 1 and Type 2 errors in hypothesis testing?**

In hypothesis testing, Type I and Type II errors represent two types of incorrect conclusions that can occur when evaluating a null hypothesis (H₀). Understanding these errors is crucial for interpreting statistical results accurately.

**Type I Error (False Positive)**

A Type I error occurs when a true null hypothesis is incorrectly rejected. This means you conclude there is an effect or difference when, in reality, none exists. It is also referred to as a false positive.

Example: In medical testing, if a diagnostic test incorrectly indicates that a healthy person has a disease, this is a Type I error.

Probability: The likelihood of committing a Type I error is denoted by α (alpha), which is the significance level of the test (e.g., 0.05).

**Type II Error (False Negative)**

A Type II error occurs when a false null hypothesis is not rejected. This means you fail to detect an effect or difference that actually exists. It is also referred to as a false negative.

Example: In medical testing, if a diagnostic test fails to detect a disease in a person who actually has it, this is a Type II error.

Probability: The likelihood of committing a Type II error is denoted by β (beta). The complement of β (1 - β) is the statistical power, which measures the test's ability to detect a true effect.

This test calculates a Z-score, which measures how many standard deviations the sample mean is away from the population mean.

**- When is the Z-Test Used?**
The Z-test is appropriate in the following situations:

**- Large sample size (usually n > 30): **The Central Limit Theorem ensures the sample mean distribution is approximately normal.

**- Known population standard deviation (σ):** The test requires the population variance or standard deviation to be known. This is often a limitation in practical scenarios.

**- Comparing a sample mean to a population mean: **You want to test if the sample mean significantly differs from the population mean.

**- Testing proportions:** The Z-test can also be used to test hypotheses about population proportions.

**10.What is the T-distribution, and when should it be used instead of the normal distribution?**

 - The T-distribution is similar in shape to the normal distribution—it is symmetric and bell-shaped—but it has heavier tails. This means that it accounts for more variability in the data, making it more suitable for small sample sizes where the assumption of normality may not hold as accurately as with larger samples. The T-distribution arises specifically under conditions where the population standard deviation is unknown and must be estimated from the sample data. As the sample size increases (and thus the degrees of freedom, which is n−1), the T-distribution approaches the normal distribution.

 **When to Use the T-Distribution Instead of the Normal Distribution**

**1.Sample Size:** The T-distribution is generally used when the sample size is small (typically n<30). When the sample sizes are larger, the Central Limit Theorem tells us that the sampling distribution of the sample mean approaches normality, so the normal distribution can be used.

**2.Unknown Population Standard Deviation: **If the population standard deviation is unknown, which is common in practical scenarios, the T-distribution should be used because it compensates for this uncertainty by incorporating a degree of variability that reflects the sample's standard deviation.

**3.Hypothesis Testing and Confidence Intervals:** The T-distribution is particularly useful in constructing confidence intervals and in performing hypothesis tests for population means when the conditions above apply. For example, if you want to estimate the mean height of students in a small class and you only collected data from 15 students, you should use the T-distribution to calculate your confidence intervals and test hypotheses.

**12.What is the T-test, and how is it used in hypothesis testing?**

  - A **T-test** (Student’s T-test) is a type of inferential statistical test used to evaluate whether the means of two groups are statistically different from each other. This method is particularly useful when the sample size is small, and the population standard deviation is unknown.

 - ** Performing a T-Test**

**1.Formulate Hypotheses:** Define the null hypothesis (H0), which asserts that the population means are equal, and the alternative hypothesis (Ha), which states they are not.

**2.Calculate the T-Statistic:** Use the formula to find the T-value, which reflects the difference between group means divided by the variability in the data.

**3.Determine Significance:** Compare the calculated T-value to a critical value from the T-distribution table at a specified significance level (commonly 0.05). If the calculated T-value is greater than the critical value, you reject the null hypothesis.

**4.Interpret Results:** Results are interpreted based on the T-value, degrees of freedom, and p-value. A small p-value (typically <0.05) indicates strong evidence against the null hypothesis.

**14.What is a confidence interval, and how is it used to interpret statistical results?**

- **A confidence interval** is calculated from sample data and provides a range of plausible values for a population parameter. For example, if you calculate a 95% confidence interval for the average height of adults in a city, it means that you are 95% confident that the true average height for all adults in that city falls within that range.

- **How it's used:**

 **- Quantifying Uncertainty:** Confidence intervals help quantify the uncertainty associated with estimates derived from sample data. A wider interval suggests more uncertainty, while a narrower interval suggests more precision.

**- Interpreting Statistical Significance:** If a confidence interval for the difference between two groups does not include zero, it suggests that the difference is statistically significant.

**- Comparing Results: **Confidence intervals can be used to compare results from different studies or samples. If the intervals overlap, it suggests that the results may not be significantly different.
Making Inferences: Confidence intervals help researchers make inferences about a population based on sample data.


**15.What is the margin of error, and how does it affect the confidence interval?**

- **The margin of error **is a statistical measure of how much the results of a survey or study might differ from the true population value. It's a range of values, typically expressed as a plus or minus percentage, that indicates the potential variability in your results. A smaller margin of error means your results are more precise and likely closer to the true population value.

- **How Margin of Error Affects Confidence Intervals:**

- Confidence intervals provide a range of values within which the true population parameter is likely to fall, given a certain level of confidence.

- The margin of error is a crucial component of the confidence interval calculation.

- It essentially determines the width of the confidence interval.

- A larger margin of error results in a wider confidence interval, while a smaller margin of error leads to a narrower confidence interval.

- For example, a 95% confidence interval with a margin of error of +/- 5% means you are 95% confident that the true population value lies within 5 percentage points above or below your sample result.

- A higher confidence level (e.g., 99%) generally requires a larger margin of error to achieve the same level of certainty, resulting in a wider interval.

- Conversely, a lower confidence level (e.g., 90%) can be achieved with a smaller margin of error, leading to a narrower interval.

- Therefore, the margin of error directly influences the precision and certainty associated with the confidence interval.

**16.How is Bayes' Theorem used in statistics, and what is its significance?**

**Bayes' Theorem in Statistics**

Bayes' theorem is expressed mathematically as:

**P(A|B)** = [P(B|A) * P(A)] / P(B)
Where:
**P(A|B)**
is the posterior probability: The probability of event A happening given that event B has already occurred.
**P(B|A)**
is the likelihood: The probability of event B happening given that event A has already occurred.
**P(A)**
is the prior probability: The initial probability of event A before considering any new evidence.
**P(B)**
is the marginal probability: The probability of event B occurring, regardless of event A.

**Significance of Bayes' Theorem**

**1. Updating Probabilities:**
Bayes' theorem is significant because it allows us to update our beliefs or probabilities as new evidence becomes available. We start with a prior probability, and as we observe new data, we can calculate a posterior probability that reflects the updated likelihood of an event.

**2. Inversion of Conditional Probabilities:**
It allows us to calculate the probability of a cause (event A) given its effect (event B), even if it's easier to directly observe the probability of the effect given the cause (B|A).

**3. Bayesian Inference:**
Bayes' theorem is foundational to Bayesian inference, a statistical approach that uses Bayes' theorem to update probabilities of hypotheses or models based on observed data.

**4. Applications in Various Fields:**
Bayes' theorem has diverse applications, including:

- Medical Diagnosis: Determining the probability of a disease given symptoms or test results.

- Spam Filtering: Classifying emails as spam based on the presence of certain words.

- Machine Learning: Used in algorithms like Naive Bayes for classification and spam filtering.

- Financial Modeling: Assessing investment risks and returns.
Risk Assessment: Evaluating the probability of various risks in different scenarios.

- Anomaly Detection: Identifying unusual patterns or outliers in datasets.
Personalization: Recommending products or content based on user preferences.

**18.What is the Chi-square goodness of fit test, and how is it applied?**

Chi Square Test  Chi-Square Test Explained Probability And Statistics  Tutorial  SimplilearnThe chi-square goodness-of-fit test is a statistical test used to determine if a sample distribution matches a known or expected distribution. It assesses how well observed frequencies of a categorical variable align with expected frequencies based on a theoretical distribution.

**Here's a breakdown of how it's applied:**

**1. Define the Hypotheses:**

- Null Hypothesis (H0): The observed data follows the expected distribution.

- Alternative Hypothesis (Ha): The observed data does not follow the expected distribution.

**2. Determine Expected Frequencies:**

- For each category in your observed data, calculate the expected frequency based on the hypothesized distribution.

- If no specific distribution is hypothesized, equal probabilities (e.g., a uniform distribution) are often assumed.

**3. Calculate the Chi-Square Statistic: **

- The chi-square statistic is calculated using the following formula: χ² = Σ [(Observed - Expected)² / Expected].

- Sum the results for each category.

**4. Determine Degrees of Freedom:**

- Degrees of freedom (df) are calculated as (number of categories - 1).

**5. Determine the p-value:**

- Using the calculated chi-square statistic and degrees of freedom, find the p-value from a chi-square distribution table or using statistical software.

- The p-value represents the probability of observing a chi-square statistic as extreme as, or more extreme than, the one calculated, if the null hypothesis is true.

**6. Make a Decision:**

- Compare the p-value to a pre-determined significance level (alpha, typically 0.05).

- If the p-value is less than or equal to alpha, reject the null hypothesis and conclude that the observed distribution is significantly different from the expected distribution.

- If the p-value is greater than alpha, fail to reject the null hypothesis and conclude that there is not enough evidence to say the observed distribution is significantly different from the expected distribution.

**20.What is an ANOVA test, and what are its assumptions?**

  - **ANOVA **(Analysis of Variance) is a statistical test used to compare the means of three or more groups. It determines if there are any statistically significant differences between the group means, indicating that the observed differences are likely not due to random chance. The validity of ANOVA results relies on several key assumptions.

**Assumptions of ANOVA: **

**1. Normality:**
The data within each group being compared should be approximately normally distributed. This means the data should follow a bell-shaped curve with most values clustered around the mean.

**2. Homogeneity of Variance (Homoscedasticity):**
The variance (spread or dispersion) of the data should be roughly equal across all the groups being compared.

**3. Independence of Observations:**
The observations within each group and between different groups should be independent of each other. This means that the value of one data point should not influence the value of another.

**21.What are the different types of ANOVA tests?**

The primary types of ANOVA tests are one-way ANOVA and two-way ANOVA.

**Here's a more detailed breakdown:**

- This is the simplest form of ANOVA and is used when you have one independent variable with three or more levels (groups) and one dependent variable.

- It determines if there are any statistically significant differences between the means of the groups.

- Example: Comparing the test scores of students from three different schools to see if there's a significant difference in their average performance.

**2. Two-way ANOVA: **

- This type of ANOVA involves two or more independent variables and one dependent variable.

- It can assess the main effects of each independent variable on the dependent variable, as well as any interaction effects between the independent variables.

- Example: Investigating the effect of both fertilizer type (independent variable 1) and watering frequency (independent variable 2) on plant growth (dependent variable).

**2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1    # inclusive lower bound
high = 6   # exclusive upper bound

# Generate the discrete uniform distribution values
x = np.arange(low, high)  # Creates an array [1, 2, 3, 4, 5]
pmf = np.ones_like(x) / len(x)  # PMF: Each value has equal probability

# Plotting the PMF
plt.bar(x, pmf, width=0.4, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function (PMF) of Discrete Uniform Distribution')
plt.xlabel('Values')
plt.ylabel('Probability')
plt.xticks(x)  # Sets x-ticks to be the values
plt.ylim(0, 1)  # Limits for the y-axis
plt.grid(axis='y', linestyle='--')
plt.show()
```

**3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pdf(x, p):
    """
    Calculate the probability distribution function of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (0 <= p <= 1)

    Returns:
    float: The probability of the outcome
    """
    if x not in [0, 1]:
        return 0  # Not a valid Bernoulli outcome
    return p if x == 1 else 1 - p

# Example usage
p = 0.7  # Probability of success
outcomes = [0, 1]  # Possible outcomes
pdf_values = {outcome: bernoulli_pdf(outcome, p) for outcome in outcomes}

print("PDF values for Bernoulli distribution with p =", p)
print(pdf_values)
```

**4.Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram**

**14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution**

```python
import scipy.stats as stats

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.
    Z = (X - μ) / σ
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution.
    Returns the probability of the data point being less than or equal to the given Z-score.
    """
    probability = stats.norm.cdf(z_score)  # Cumulative distribution function
    return probability

# Example usage
data_point = 75
mean = 70
std_dev = 5

z_score = calculate_z_score(data_point, mean, std_dev)
probability = compare_to_standard_normal(z_score)

print(f"Z-score: {z_score:.2f}")
print(f"Probability (P(Z ≤ {z_score:.2f})): {probability:.4f}")
```

**24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.

    Parameters:
        data (array-like): Input dataset (e.g., list or numpy array).

    Returns:
        tuple: Skewness and kurtosis of the dataset.
    """
    skewness = skew(data)
    kurt = kurtosis(data, fisher=True)  # Fisher=True gives excess kurtosis
    return skewness, kurt

# Example usage:
if __name__ == "__main__":
    # Generate a dataset from a normal distribution
    np.random.seed(42)  # For reproducibility
    normal_data = np.random.normal(loc=0, scale=1, size=1000)

    skewness, kurt = calculate_skewness_kurtosis(normal_data)
    print(f"Skewness: {skewness}")
    print(f"Kurtosis (Excess): {kurt}")
```

**1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results**

```python
import scipy.stats as stats
import math

# Function to perform Z-test
def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Interpretation of results
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Since p-value ({p_value:.4f}) < alpha ({alpha}), we reject the null hypothesis.")
        print("Interpretation: The sample mean is significantly different from the population mean.")
    else:
        print(f"Since p-value ({p_value:.4f}) >= alpha ({alpha}), we fail to reject the null hypothesis.")
        print("Interpretation: The sample mean is not significantly different from the population mean.")

# Example usage
# Known values
population_mean = 50  # Population mean
population_std = 10   # Population standard deviation
sample_mean = 53      # Sample mean
sample_size = 30      # Sample size
alpha = 0.05          # Significance level

# Perform the Z-test
z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

**3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean**

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample: List or array of sample data.
    - population_mean: The mean of the population.
    - population_std: The standard deviation of the population.

    Returns:
    - z_score: The calculated Z-score.
    - p_value: The p-value corresponding to the Z-score.
    """
    # Calculate the sample mean
    sample_mean = np.mean(sample)

    # Calculate the sample size
    n = len(sample)

    # Calculate the standard error
    standard_error = population_std / np.sqrt(n)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / standard_error

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    return z_score, p_value

# Example usage
sample_data = [50, 52, 48, 49, 51, 50, 53, 47]
population_mean = 50
population_std = 5

z_score, p_value = one_sample_z_test(sample_data, population_mean, population_std)

print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")
```

**4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest
# Sample data
sample1 = np.random.normal(loc=50, scale=10, size=100)  # Sample 1
sample2 = np.random.normal(loc=52, scale=10, size=100)  # Sample 2
# Perform two-sample Z-test
z_statistic, p_value = ztest(sample1, sample2, value=0)  # H0: mean1 - mean2 = 0
print(f"Z-Statistic: {z_statistic}, P-Value: {p_value}")
# Define significance level
alpha = 0.05

# Z critical values for two-tailed test
z_critical = stats.norm.ppf(1 - alpha / 2)

# Create a range of values for the X-axis
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, loc=0, scale=1)  # Standard normal distribution

# Plotting
plt.figure(figsize=(10, 6))

# Plot the normal distribution
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, 0, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label='Rejection region')
plt.axvline(z_critical, color='gray', linestyle='--')
plt.axvline(-z_critical, color='gray', linestyle='--')

# Mark the Z-statistic on the plot
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Z-Statistic: {z_statistic:.2f}')
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

**5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def visualize_type_errors(alpha=0.05, effect_size=1, sample_size=30, num_simulations=1000):
    # Calculate critical value for Type I error
    critical_value = stats.norm.ppf(1 - alpha)

    # Store results
    type_1_errors = 0
    type_2_errors = 0
    results = []

    # Simulation
    for _ in range(num_simulations):
        # Generate sample data from null (H0: mu=0) and alternative hypothesis (H1: mu=effect_size)
        null_sample = np.random.normal(0, 1, sample_size)  # H0: mean = 0
        alternative_sample = np.random.normal(effect_size, 1, sample_size)  # H1: mean = effect_size

        # Perform t-test
        t_stat, p_value = stats.ttest_ind(null_sample, alternative_sample, equal_var=False)

        # Checking for Type I Error
        if p_value < alpha:  # Reject H0 when it is actually true
            type_1_errors += 1  # Type I Error occurred

        # Checking for Type II Error
        t_stat_alt, p_value_alt = stats.ttest_1samp(alternative_sample, 0)  # Test against H0: mean = 0
        if p_value_alt > alpha:  # Fail to reject H0 when it's false
            type_2_errors += 1  # Type II Error occurred

        results.append((t_stat, p_value, p_value_alt))

    # Calculate Type I and Type II error rates
    type_1_error_rate = type_1_errors / num_simulations
    type_2_error_rate = type_2_errors / num_simulations

    # Visualization
    plt.figure(figsize=(10, 6))
    sns.histplot([result[0] for result in results], bins=30, kde=True, color='blue', label='T-Statistic under H0')
    plt.axvline(x=critical_value, color='red', linestyle='--', label='Critical Value (alpha)')
    plt.axvline(x=-critical_value, color='red', linestyle='--')
    plt.title('Distribution of T-Statistics')
    plt.xlabel('T-Statistic')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

    return type_1_error_rate, type_2_error_rate

# Example usage
alpha_value = 0.05
effect_size_value = 1
sample_size_value = 30
num_simulations_value = 1000

type_1, type_2 = visualize_type_errors(alpha_value, effect_size_value, sample_size_value, num_simulations_value)
print(f"Type I Error Rate: {type_1:.4f}")
print(f"Type II Error Rate: {type_2:.4f}")
```

**8.Simulate data and perform both Z-test and T-test, then compare the results using Python**

```python
import numpy as np
from scipy.stats import ttest_1samp, norm

# Simulate data
np.random.seed(42)  # For reproducibility
population_mean = 50
population_std = 10
sample_size = 30
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Known population parameters
known_population_mean = 50
known_population_std = 10

# Perform Z-test
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # Sample standard deviation
z_statistic = (sample_mean - known_population_mean) / (known_population_std / np.sqrt(sample_size))
z_p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Perform T-test
t_statistic, t_p_value = ttest_1samp(sample, popmean=known_population_mean)

# Compare results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("\nZ-Test Results:")
print(f"Z-Statistic: {z_statistic:.4f}, P-Value: {z_p_value:.4f}")
print("\nT-Test Results:")
print(f"T-Statistic: {t_statistic:.4f}, P-Value: {t_p_value:.4f}")

# Interpretation
if z_p_value < 0.05:
    print("\nZ-Test: Reject the null hypothesis (significant difference).")
else:
    print("\nZ-Test: Fail to reject the null hypothesis (no significant difference).")

if t_p_value < 0.05:
    print("T-Test: Reject the null hypothesis (significant difference).")
else:
    print("T-Test: Fail to reject the null hypothesis (no significant difference).")
```

**9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance**

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or array-like): The sample data.
        confidence (float): The confidence level (default is 0.95 for 95% confidence).

    Returns:
        tuple: A tuple containing the lower bound and upper bound of the confidence interval.
    """
    # Convert data to a numpy array for easier calculation
    data = np.array(data)

    # Calculate the sample mean
    mean = np.mean(data)

    # Calculate the sample standard deviation
    std_dev = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation

    # Calculate the sample size
    n = len(data)

    # Calculate the standard error
    standard_error = std_dev / np.sqrt(n)

    # Calculate the t-score for the confidence interval
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate the margin of error
    margin_of_error = t_score * standard_error

    # Calculate the confidence interval
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound
```

**10.Write a Python program to calculate the margin of error for a given confidence level using sample data**

```python
import math
from scipy.stats import norm

def calculate_margin_of_error(confidence_level, standard_deviation, sample_size):
    # Calculate the critical z-value for the given confidence level
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the standard error
    standard_error = standard_deviation / math.sqrt(sample_size)

    # Calculate the margin of error
    margin_of_error = z_score * standard_error

    return margin_of_error

# Example usage
confidence_level = 0.95  # 95% confidence level
standard_deviation = 15  # Example standard deviation
sample_size = 100        # Example sample size

margin_of_error = calculate_margin_of_error(confidence_level, standard_deviation, sample_size)
print(f"Margin of Error: {margin_of_error:.2f}")
```

**11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process**

```python
def bayes_theorem(p_a, p_b_given_a, p_b):
    """Apply Bayes' Theorem to calculate the posterior probability."""
    return (p_b_given_a * p_a) / p_b

# Define the probabilities
p_a = 0.2             # Probability of rain
p_b_given_a = 0.85    # Probability of clouds given it's raining
p_b = 0.4             # Probability of clouds

# Use Bayes' Theorem
p_rain_given_clouds = bayes_theorem(p_a, p_b_given_a, p_b)
print(f'The probability that it will rain given that it is cloudy is: {p_rain_given_clouds:.2f}')
```

**12.Perform a Chi-square test for independence between two categorical variables in Python**

```python
import pandas as pd
from scipy.stats import chi2_contingency
data = {'Preference': ['Yes', 'Yes', 'No', 'No'],
        'Gender': ['Male', 'Female', 'Male', 'Female']}
df = pd.DataFrame(data)

contingency_table = pd.crosstab(df['Gender'], df['Preference'])
print(contingency_table)
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square Statistic: {chi2}, p-value: {p}, Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(expected)
if p < 0.05:
    print("Reject the null hypothesis: The variables are associated.")
else:
    print("Fail to reject the null hypothesis: The variables are independent.")
```

**13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data**

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
        observed (2D list or numpy array): Observed frequency table.

    Returns:
        numpy array: Expected frequency table.
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()

    # Calculate expected frequencies
    expected = np.outer(row_totals, col_totals) / grand_total
    return expected

# Example usage
observed_data = [
    [50, 30, 20],
    [30, 40, 30],
    [20, 30, 50]
]

expected_frequencies = calculate_expected_frequencies(observed_data)
print("Observed Data:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

**14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.**

```python
import numpy as np
from scipy.stats import chisquare

# Observed data
observed = np.array([50, 30, 20])

# Expected data (must sum to the same total as observed)
expected = np.array([40, 40, 20])

# Perform the Chi-Square Goodness-of-Fit Test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("The observed data does NOT fit the expected distribution (reject null hypothesis).")
else:
    print("The observed data fits the expected distribution (fail to reject null hypothesis).")
```
